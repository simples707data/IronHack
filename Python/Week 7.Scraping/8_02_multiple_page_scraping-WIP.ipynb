### putting everything together 

```python
#we are building everything from scratch here 

pages= []

for i in iterations: 
    #takes the iteration, turns it into a string (not needed)
    start_at = str(i)
    #define the url
    url= f"https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start={start_at}&ref_=adv_nxt"
    # download the url content 
    response = requests.get(url)
    #print status code 
    print("Status code: "+ str(response.status_code))
    #to capture the data and fill in the list
    pages.append(response)
    #add a respectful nap 
    wait_time=randint(1,4)
    sleep(wait_time)
```

```python
soup=BeautifulSoup(pages[0].content, "html.parser")
```
